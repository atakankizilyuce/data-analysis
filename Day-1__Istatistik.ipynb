# Create simple and custom Dataset

Let's create a Dataset that has a hundred students in it. Each student has a gender, faculty, GPA, and id.

```python
import random
import pandas as pd
import numpy as np

# Generating Gender, Faculties, and GPA Columns for the custom Dataset
genders = []
faculties = []
gpa = []
id = []

for i in range(100):
  genders.append(random.choice(["female","male"]))
  faculties.append(random.choice(["COE", "EEE", "IE"]))
  gpa.append(random.uniform(0, 4))
  id.append(random.randint(0000000,9999999))


# Creating a dictionary to store all of the generated values together
dictionary_form = {'gender': genders,                   
                   'faculty': faculties,
                   'gpa': gpa,
                   'id':id}

# Converting the Dictionary to Dataframe format to be read by Pandas
dataframe_form = pd.DataFrame(dictionary_form)
dataframe_form.head()
```

```python
# Generating Gender, ID and grades of Data Structures, Linear Algebra, Advanced Programming, Database Systems, Cyber Security, Computer Vision for the custom Dataset
genders = []
id = []
data_structures = []
linear_algebra = []
advanced_programming = []
database_systems = []
cyber_security = []
computer_vision = []


for i in range(100):
  genders.append(random.choice(["female","male"]))
  data_structures.append(random.randint(0,100))
  linear_algebra.append(random.randint(0,100))
  advanced_programming.append(random.randint(0,100))
  database_systems.append(random.randint(0,100))
  cyber_security.append(random.randint(0,100))
  computer_vision.append(random.randint(0,100))
  # to make id unique
  while True:
    student_id = random.randint(1000000,9999999)
    if not student_id in id:
      id.append(student_id)
      break
# Creating a dictionary to store all of the generated values together
dictionary_form = {'Gender': genders, 
                  'Data Structures': data_structures,
                  'Linear Algebra': linear_algebra,
                  'Advanced Programming': advanced_programming,
                  'Database Systems': database_systems,
                  'Cyber Security': cyber_security,
                  'Computer Vision': computer_vision,
                  'ID':id}

# Converting the Dictionary to Dataframe format to be read by Pandas
df = pd.DataFrame(dictionary_form)
df.head()
```

```python
# Configure plot settings
plt.figure(figsize=(15,10))
plt.xticks([])
plt.title("Grade-Student")
plt.xlabel("Student")
plt.ylabel("Grade")

# Plot grades
plt.scatter(df["Data Structures"].index,df["Data Structures"])

# Draw mean line
plt.axhline(df["Data Structures"].mean(),linestyle ="dashed",color="black")
plt.show()
```

```python
print(df["Data Structures"].mean())
print(df["Linear Algebra"].mean())
```

```python
if df["Data Structures"].mean() < 60:
  df["Data Structures"] = df["Data Structures"] + (60-df["Data Structures"].mean())
  df.loc[df["Data Structures"] > 100,"Data Structures"] = 100

if df["Linear Algebra"].mean() < 60:
  df["Linear Algebra"] = df["Linear Algebra"] + (60-df["Linear Algebra"].mean())
  df.loc[df["Linear Algebra"] > 100,"Linear Algebra"] = 100
```
